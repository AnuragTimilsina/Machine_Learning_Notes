```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
 from sklearn.cluster import KMeans
```

```python
stars = pd.read_csv('cleaned_stars.csv')
```

```python
stars.sample(10)
```

```python
stars.head()
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
stars['Colour'] = LE.fit_transform(stars['Colour'])
stars
```

```python
LE = LabelEncoder()
stars['Type'] = LE.fit_transform(stars['Type'])
stars
```

```python
LE = LabelEncoder()
stars['Spectral_Class'] = LE.fit_transform(stars['Spectral_Class'])
stars
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(stars)
```

```python
kmeans.labels_
```

```python
distortions = []
K = range(1, 15)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(stars)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
