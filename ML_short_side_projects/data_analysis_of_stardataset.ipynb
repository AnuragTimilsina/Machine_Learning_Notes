```python
# Function to get the lower and upper bound:
```

```python
def get_upper_lower(df, column_name):
  q1 = df[column_name].quantile(0.25)
  q3 = df[column_name].quantile(0.75)
  iqr = q3 - q1
  lower_limit = q1 - 1.5*iqr
  upper_limit = q3 + 1.5*iqr
  return upper_limit, lower_limit
```

```python
column_name = 'Temperature'
upper, lower = get_upper_lower(cleaned_df, column_name)
```

```python
upper, lower
```

```python
cleaned_df = cleaned_df[(cleaned_df.Temperature>lower)&(cleaned_df.Temperature<upper)]
```

```python
column_name = 'Luminosity'
upper, lower = get_upper_lower(cleaned_df, column_name)
```
